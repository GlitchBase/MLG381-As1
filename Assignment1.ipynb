```python
import numpy as np
import pandas as pd
import statistics
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('auto.csv')
df.dropna(how='any')
df
```

```python
mpg01 = 0
dfmean = df['mpg'].mean()
dfmean
```

```python
Values = []

for index, row in df.iterrows():
    if row['mpg'] > dfmean :
        mpg01 = 1
    else:
        mpg01 = 0
    Values.append([row['mpg'],mpg01,row['cylinders'],row['horsepower'],row['year']])
    
    #print(index,'\t',row['mpg'],'\t',mpg01,'\t',row['cylinders'],'\t',row['horsepower'],'\t',row['weight'],'\t',row['year'])
dfCylComp = pd.DataFrame(Values,columns=['mpg','MPG01','Cylinders','Horsepower','Year'])
dfCylComp 
```

```python
Values2=[]
for index, row in df.iterrows():
    if row['mpg'] > dfmean :
        mpg01 = 1
    else:
        mpg01 = 0
    Values2.append([row['mpg'],mpg01,row['horsepower'],row['acceleration']])
    
dfCalc = pd.DataFrame(Values2,columns=['mpg','MPG01','Horsepower','Acceleration'])
print(dfCalc)
dfCalc.plot()
plt.show()
```
